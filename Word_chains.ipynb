```python
import networkx as nx
from networkx.drawing.nx_pydot import graphviz_layout
import matplotlib.pyplot as plt
from num2words import num2words
```

```python
DG = nx.DiGraph()
```

```python
def next_num(n:int) -> int:
  next = len(num2words(n, lang='es').replace(' ', '').replace('-', ''))  # Change 'lang' kwarg to draw graph for different languages
  return next

DG.clear()
for i in range(0, 101):
  n = i
  next = next_num(n)
  DG.add_edge(n, next)
  while next not in DG.nodes:
    n = next
    next = next_num(n)
    DG.add_edge(n, next)

pos = graphviz_layout(DG, prog="circo")
options = {
    "with_labels":True,
    "font_size": 10,
    "node_size": 300,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 1,
    "width": 1,
}
 
plt.figure(figsize=(20,10)) 
nx.draw(DG, pos, **options)
plt.show()
```
